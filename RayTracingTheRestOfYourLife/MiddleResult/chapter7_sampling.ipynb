```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 500
pi = 3.1415926
pos_x = np.empty([1,N], dtype = float)
pos_y = np.empty([1,N], dtype = float)
pos_z = np.empty([1,N], dtype = float)
# code as RTleft tutorial bool



for i in range(N):
    r1 = np.random.uniform(0,1)
    r2 = np.random.uniform(0,1)
    pos_x[0,i] = np.cos(2 * pi * r1) * 2 * np.sqrt(r2 * (1 - r2))
    pos_y[0,i] = np.sin(2 * pi * r1) * 2 * np.sqrt(r2 * (1 - r2))
    pos_z[0,i] = (1 - 2 * r2)


fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(pos_x,pos_y,pos_z,  cmap='pink_r')
plt.show()
```

```python
# hemisphere
N = 2000
pi = 3.1415926
pos_x = np.empty([1,N], dtype = float)
pos_y = np.empty([1,N], dtype = float)
pos_z = np.empty([1,N], dtype = float)
# code as RTleft tutorial bool



for i in range(N):
    r1 = np.random.uniform(0,1)
    r2 = np.random.uniform(0,1)
    pos_x[0,i] = np.cos(2 * pi * r1) * np.sqrt(r2)
    pos_y[0,i] = np.sin(2 * pi * r1) * np.sqrt(r2)
    pos_z[0,i] = np.sqrt(1 - r2)


fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(pos_x,pos_y,pos_z,  cmap='pink_r')
ax.set_zlim(0, 1.5)
plt.show()
```
